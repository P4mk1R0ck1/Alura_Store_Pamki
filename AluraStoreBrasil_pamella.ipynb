### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
faturamento_loja = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print("Faturamento Loja 1:", faturamento_loja)
print("Faturamento Loja 2:", faturamento_loja2)
print("Faturamento Loja 3:", faturamento_loja3)
print("Faturamento Loja 4:", faturamento_loja4)
```

```python
faturamentos = {
    'Loja 1': faturamento_loja,
    'Loja 2': faturamento_loja2,
    'Loja 3': faturamento_loja3,
    'Loja 4': faturamento_loja4
}

plt.figure()
plt.bar(faturamentos.keys(), faturamentos.values())

plt.title('Faturamento Total por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento')

plt.show()
```

###Gráfico de Dispersão —> **Avaliação vs Frete Médio**

```python
avaliacoes = [
    media_avaliacao_loja,
    media_avaliacao_loja2,
    media_avaliacao_loja3,
    media_avaliacao_loja4
]

fretes = [
    frete_medio_loja1,
    frete_medio_loja2,
    frete_medio_loja3,
    frete_medio_loja4
]

nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
```

```python
plt.figure()
plt.scatter(fretes, avaliacoes)

for i in range(len(nomes_lojas)):
    plt.text(fretes[i], avaliacoes[i], nomes_lojas[i])

plt.xlabel('Frete Médio')
plt.ylabel('Média de Avaliação')
plt.title('Relação entre Frete Médio e Avaliação')

plt.show()
```

#RELATÓRIO FINAL – Análise das Lojas da Rede Alura Store

Este relatório tem como objetivo analisar o desempenho das quatro lojas da rede Alura Store para apoiar a decisão do Senhor João sobre qual unidade vender.

Foram avaliados indicadores financeiros, operacionais e de satisfação do cliente, buscando entender qual loja apresenta o desempenho menos equilibrado.

### Análise dos Dados

**Faturamento**

A comparação do faturamento total permitiu identificar diferenças no desempenho financeiro entre as lojas. Esse indicador mostra a capacidade de geração de receita e é essencial para avaliar a sustentabilidade do negócio.

Também foram analisadas as categorias mais vendidas e os produtos com maior e menor saída, permitindo entender o perfil de consumo e a força comercial de cada unidade.

**Avaliação dos Clientes**

A média das avaliações apresentou os seguintes resultados:

**Loja 1: 3,97**

**Loja 2: 4,03**

**Loja 3: 4,04**

**Loja 4: 3,99**

**A Loja 3** apresentou a melhor média de satisfação, enquanto a **Loja 1** teve a menor avaliação entre as quatro unidades.

Embora as diferenças não sejam extremamente altas, a avaliação é um indicador importante de experiência do cliente e pode impactar diretamente a fidelização e a reputação da loja.

**Frete Médio**

O frete médio apresentou os seguintes valores:

**Loja 1: 34,69**

**Loja 2: 33,62**

**Loja 3: 33,07**

**Loja 4: 31,27**

A **Loja 1** possui o frete médio mais alto, enquanto a **Loja 4** apresenta o menor custo logístico.

Como o frete é pago pelo cliente, valores mais elevados podem reduzir a competitividade e influenciar negativamente a experiência de compra.

**Concluindo**

Ao cruzar os principais indicadores, observa-se que a **Loja 1 apresenta o menor equilíbrio** entre satisfação do cliente e competitividade logística, pois combina:

**O frete médio mais alto entre as lojas;**

**A menor média de avaliação dos clientes.**

Esse cenário indica possível impacto do custo logístico na experiência do consumidor, o que pode comprometer a competitividade da unidade no longo prazo.

Dessa forma, considerando o conjunto das análises realizadas, recomenda-se que o Senhor João avalie a **venda da Loja 1**, mantendo as demais unidades que demonstraram melhor equilíbrio entre desempenho financeiro, satisfação dos clientes e eficiência logística.
